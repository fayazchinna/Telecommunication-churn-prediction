```python
#numerical and catagorical plot with function
```

```python
def plot_numerical_data(df):
  for col in df.columns:
    plt.figure(figsize=(10,5))
    sns.histplot(df[col], kde=True,bins=10,color='skyblue')
    plt.title(f'distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plot_numerical_data(df.select_dtypes(include=['number']))
```

```python
def plot_catagorical_data(df):
  for col in df.columns:
    plt.figure(figsize=(16,8))
    sns.countplot(x=col, data=df,palette='Set2')
    plt.title(f'distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.show()
```

```python
plot_catagorical_data(df.select_dtypes(include=['object', 'category']))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Churn', axis=1), df['Churn'], test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
#apply standred scaler in dataset
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train.select_dtypes(include=np.number))
x_train = scaler.transform(X_train.select_dtypes(include=np.number))
x_test = scaler.transform(X_test.select_dtypes(include=np.number))
```

```python
x_train.shape, x_test.shape
```

```python
#MODEL BUILDING
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
log_reg = LogisticRegression(C=0.001, penalty='l2', solver='liblinear')
log_reg.fit(x_train, y_train)
```

```python
predlr = log_reg.predict(x_test)
```

```python
#GRID SERCH TO FIND BEST PARAMS
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}

# Create GridSearchCV object
grid_search = GridSearchCV(estimator=log_reg, param_grid=param_grid, cv=5, scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(x_train, y_train)

# Get the best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)

# Use the best model for predictions
best_model = grid_search.best_estimator_
predictions = best_model.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, roc_curve, auc
import matplotlib.pyplot as plt

# Create the Random Forest model
model = RandomForestClassifier(n_estimators=400, n_jobs=-1, random_state=0)

# KFold for cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation and get scores
cv_scores = cross_val_score(model, x_train, y_train, cv=kf, scoring='accuracy')

# Print cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", cv_scores.mean())

# Train the model on the entire training data for final evaluation
model.fit(x_train, y_train)

# Store trained model
trained_rf_model = model

# Get predictions and probabilities
rf_pred = trained_rf_model.predict(x_test)
rf_pred_proba = trained_rf_model.predict_proba(x_test)[:, 1]

# Calculate accuracy and AUC
rf_acc = accuracy_score(y_test, rf_pred)
rf_auc = roc_auc_score(y_test, rf_pred_proba)
print("Random Forest Accuracy:", rf_acc)
print("Random Forest AUC:", rf_auc)

# Print classification report
print(classification_report(y_test, rf_pred))

# Plot ROC AUC curve
fpr, tpr, thresholds = roc_curve(y_test, rf_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Random Forest')
plt.legend(loc="lower right")
plt.show()
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, roc_curve, auc
```

```python
dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test, label=y_test)
xgb_params = {
    'objective': 'binary:logistic',                      #as binary classification using logistic regression. This is appropriate for predicting churn (a binary outcome).
    'max_depth': 3,
    'eta': 0.1,
    'eval_metric': 'auc'
}

xgb_model = xgb.train(xgb_params, dtrain, num_boost_round=100)
xgb_pred = xgb_model.predict(dtest)
xgb_pred_proba = xgb_model.predict(dtest)
xgb_pred = (xgb_pred > 0.5).astype(int)  #This line converts the predicted probabilities into binary predictions (0 or 1) using a threshold of 0.5. If the predicted probability is greater than 0.5, it's classified as churn (1); otherwise, it's classified as no churn (0).
xgb_acc = accuracy_score(y_test, xgb_pred)  #Calculates the accuracy of the model's predictions by comparing them to the actual churn values (y_test).
print("XGBoost Accuracy:", xgb_acc)
xgb_auc = roc_auc_score(y_test, xgb_pred) #Calculates the AUC, a metric that measures the model's ability to distinguish between churned and non-churned customers. A higher AUC indicates better performance.
print("XGBoost AUC:", xgb_auc)

# Print classification report
print(classification_report(y_test, xgb_pred))

# Plot ROC AUC curve
fpr, tpr, thresholds = roc_curve(y_test, xgb_pred_proba)
roc_auc = auc(fpr, tpr)


plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()


```

```python

import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, roc_curve, auc
import matplotlib.pyplot as plt

# Define the ANN model
model = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(x_train.shape[1],)), # Input layer with 128 neurons and ReLU activation
    keras.layers.Dense(64, activation='relu'),  # Hidden layer with 64 neurons and ReLU activation
    keras.layers.Dense(1, activation='sigmoid')  # Output layer with 1 neuron and sigmoid activation for binary classification
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',  # Loss function for binary classification
              metrics=['accuracy', 'AUC'])  # Metrics to track during training

# Train the model
history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
y_pred_proba = model.predict(x_test)
y_pred = (y_pred_proba > 0.5).astype(int)  # Convert probabilities to binary predictions

# Calculate metrics
ann_acc = accuracy_score(y_test, y_pred)
ann_auc = roc_auc_score(y_test, y_pred_proba)
print("ANN Accuracy:", ann_acc)
print("ANN AUC:", ann_auc)

# Print classification report
print(classification_report(y_test, y_pred))

# Plot ROC AUC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - ANN')
plt.legend(loc="lower right")
plt.show()
```
